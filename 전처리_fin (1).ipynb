```python
import pandas as pd
import re

df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Recipe/만개의_레시피.xlsx')
df = df.drop(columns=['RCMM_CNT', 'SRAP_CNT', 'CKG_STA_ACTO_NM', 'CKG_MTRL_ACTO_NM', 'CKG_KND_ACTO_NM'])
df.head(5)
```

```python
new_column = ['요리명', '조회수', '조리방법', '재료', '인분수', '난이도', '조리시간']

df.columns = new_column
df = df.dropna()
df['재료'] = df['재료'].str.replace(r'^\[재료\] ', '', regex=True)
df = df[df['인분수'] == '1인분']

sorted_df = df.sort_values(by='요리명')
sorted_df = sorted_df.reset_index(drop=True)
sorted_df
```

```python
# 이상한 문자를 제거하는 정규 표현식 패턴
pattern = r'[^\w\s가-힣]'
df_cleaned = sorted_df[~sorted_df['요리명'].str.contains(pattern, regex=True)]
df_cleaned
```

```python
df_cleaned = df_cleaned.loc[df_cleaned.groupby('요리명')['조회수'].idxmax()]
df_cleaned['재료'] = df_cleaned['재료'].apply(lambda x: re.sub(r'\[.*?\]|\?', '', x).strip())
df_cleaned['재료'] = df_cleaned['재료'].fillna('')
df_cleaned = df_cleaned[~df_cleaned['재료'].str.startswith('자세한 재료는')]

# '재료' 컬럼에서 '계란'이 포함된 문자열을 '달걀'로 변경
df_cleaned['재료'] = df_cleaned['재료'].fillna('').apply(lambda x: x.replace('계란', '달걀'))

# '요리명' 컬럼에서 '계란'이 포함된 문자열을 '달걀'로 변경 (선택 사항, 필요 시 추가)
df_cleaned['요리명'] = df_cleaned['요리명'].fillna('').apply(lambda x: x.replace('계란', '달걀'))
```

```python
df_cleaned = df_cleaned.loc[df_cleaned.groupby('요리명')['조회수'].idxmax()]
df_cleaned['재료'] = df_cleaned['재료'].apply(lambda x: re.sub(r'\[.*?\]|\?', '', x).strip())
df_cleaned['재료'] = df_cleaned['재료'].fillna('')
df_cleaned = df_cleaned[~df_cleaned['재료'].str.startswith('자세한 재료는')]

# '재료' 컬럼에서 '계란'이 포함된 문자열을 '달걀'로 변경
df_cleaned['재료'] = df_cleaned['재료'].fillna('').apply(lambda x: x.replace('계란', '달걀'))

# '요리명' 컬럼에서 '계란'이 포함된 문자열을 '달걀'로 변경 (선택 사항, 필요 시 추가)
df_cleaned['요리명'] = df_cleaned['요리명'].fillna('').apply(lambda x: x.replace('계란', '달걀'))

df_cleaned.to_excel('레시피_전처리_최종.xlsx', index=False, engine='openpyxl')
```

```python
def in_ingredient(ingredients_text):
  ingre = ['콩나물', '팽이버섯', '깻잎', '표고버섯', '사과', '베이컨', '양배추', '오징어', '당근',
         '치즈', '닭', '부추', '오이', '오리고기', '달걀', '가지', '마늘', '대파', '김치', '새송이버섯',
         '상추', '우유', '숙주', '배추', '양파', '파프리카', '파스타', '배', '고추', '돼지고기','소고기','쇠고기', '감자',
         '메추리알', '무', '라면', '떡', '새우', '스팸', '애호박', '고구마', '두부','참치캔', '참치 캔', '미나리', '식빵']

  ingre2=['오리','빵']

  ingredient=''

  for item in ingre:
    if item in ingredients_text:
      ingredient=ingredient+item+' '

  ingredients=ingredients_text.split(' ')
  for ing in ingredients:
    for item in ingre2:
      if ing==item:
        ingredient=ingredient+item+' '
  return ingredient
```

```python
df['체크할 재료'] = df['재료'].apply(lambda x: in_ingredient(str(x)))
df.head()
```

```python
import pandas as pd

df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Recipe/레시피_전처리_재료.csv')

# 특정 문자열이 포함된 행 제외
keywords = ['는법', '손질', '보관','깍기','자르기']
pattern = '|'.join(keywords)

filtered_df = df[~df['요리명'].str.contains(pattern, regex=True)]

filtered_df
```

```python

filtered_df['체크할 재료'] = filtered_df['체크할 재료'].replace({'오리고기': '오리'}, regex=True)
filtered_df['체크할 재료'] = filtered_df['체크할 재료'].replace({'식빵': '빵'}, regex=True)
filtered_df['체크할 재료'] = filtered_df['체크할 재료'].replace({'참치 캔': '참치캔'}, regex=True)
filtered_df['체크할 재료'] = filtered_df['체크할 재료'].replace({'쇠고기': '소고기'}, regex=True)
filtered_df
```
